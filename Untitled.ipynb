```python
def dictplot(d, ax=None, **kwargs):
    if ax is None:
        ax = plt
    x,y = zip(*d.items())
    ax.loglog([T_MAX_TO_INFER/_x for _x in x], y, **kwargs)
```

```python
#plt.figure(figsize=(12,3))
#fig,axs = plt.subplots(1, 3, sharey=True)
#plt.figure()
fig,axs = plt.subplots(3, 1, sharey=True,figsize=(6,10))
Y_MAX = 10
Y_MIN = 10e-6
for ax,scheme in zip(axs,results):
    print(scheme)
    for result in results[scheme]:
        dictplot(result.errors, ax=ax,
                 label=f'{result.dt_data:0.5f}', marker='o')
        ax.vlines(result.dt_data, Y_MIN, Y_MAX, linestyle='dotted')
    dictplot(baselines[scheme], ax=ax)
    ax.legend()
plt.ylim(Y_MIN, Y_MAX)
plt.show()
```
